```
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np


Max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
Min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

months = np.array((range(1,13)))


def function(x,a,k,d,c):
  return a*np.cos(k*x + d) + c

Max_param,_= curve_fit(function,months,Max)
Max_a,Max_k,Max_d,Max_c = Max_param
# print(Max_param)

fitted_max_temp = function(np.linspace(1,12,100),Max_a,Max_k,Max_d,Max_c)
plt.plot(np.linspace(1,12,100),fitted_max_temp,label = 'Fitted Max')

Min_param,_= curve_fit(function,months,Min)
Min_a,Min_k,Min_d,Min_c = Min_param
# print(Min_param)

fitted_min_temp = function(np.linspace(1,12,100),Min_a,Min_k,Min_d,Min_c)
plt.plot(np.linspace(1,12,100),fitted_min_temp,label = 'Fitted Min')

plt.scatter(months,Max,label = 'max')
plt.scatter(months,Min, label = 'min')
plt.legend()
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.show()
```
